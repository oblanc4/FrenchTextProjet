```python
import pandas as pd

def read_word_list(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return [line.strip() for line in file if line.strip()]

def calculate_match_percentage(sentence, word_list):
    """
    Calculate the percentage of words in the sentence that are also in the word list.

    Args:
    sentence (str): The sentence to analyze.
    word_list (list): The list of words to compare against.

    Returns:
    float: The percentage of words in the sentence that are in the word list.
    """    
    # Tokenize the sentence into words
    words_in_sentence = sentence.split()

    # Count how many words are in the word list
    match_count = sum(word in word_list for word in words_in_sentence)

    # Calculate the percentage
    if len(words_in_sentence) > 0:
        return (match_count / len(words_in_sentence)) * 100
    else:
        return 0
word_lists = {

    'A1': read_word_list('A1.txt'),
    'A2': read_word_list('A2.txt'),
    'B1': read_word_list('B1.txt'),
    'B2': read_word_list('B2.txt'),
    'C1': read_word_list('C1.txt'),
    'C2': read_word_list('C2.txt')
}

# Path to the CSV file
csv_file_path = 'training_data_processed.csv'  

# Read the CSV file
training_data = pd.read_csv(csv_file_path)

# Your cleaned words list from df_A1
with open('A1.txt', 'r') as file:
    cleaned_words = [line.strip() for line in file if line.strip()]

# Apply the function to each processed sentence in the DataFrame
for level, words in word_lists.items():
    training_data[f'match_percentage_{level}'] = training_data['processed_sentence'].apply(
        lambda sentence: calculate_match_percentage(sentence, set(words))
    )
# Display the first few rows of the DataFrame with the new column
columns_to_display = ['processed_sentence'] + [f'match_percentage_{level}' for level in word_lists.keys()]
training_data[columns_to_display]
```
